```python
import os
import pickle
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
location = "/content/Time-Series-Forecasting-Using-the-Bayesian-Approach-and-Random-Fourier-Features/Experiments/"

# Changue *.pkl
with open('{location}/results_proof2.pkl'.format(location=location), 'rb') as f:
    results = pickle.load(f)
```

```python
plt.figure(figsize=(20,5))
for k in range(len(results)):
  plt.subplot(1, 3, k+1)
  ax = sns.heatmap(data=results[k], annot=False, cbar=True, #cmap='Blues',
                     yticklabels=range(1, 21, 1), xticklabels=range(3, 21, 1),
                     annot_kws={"size": 14})  # Tamaño de los números dentro de la imagen
  ax.set_xlabel("Window", fontsize=18)
  ax.set_ylabel("Kernel size", fontsize=18)
  ax.set_title("Serie {}".format(k+1), fontsize=20)
plt.show()

```

```python
for k in range(len(results)):
    plt.figure(figsize=(20, 20))
    ax = sns.heatmap(data=results[k], annot=True, cbar=True, cmap='Blues',
                     yticklabels=range(1, 21, 1), xticklabels=range(3, 21, 1),
                     annot_kws={"size": 14})  # Tamaño de los números dentro de la imagen

    # Ajuste del tamaño de fuente de los ejes y título
    ax.set_xlabel("Window", fontsize=18)
    ax.set_ylabel("Kernel size", fontsize=18)
    ax.set_title("Serie {}".format(k + 1), fontsize=20)

    # Ajuste del tamaño de los tick labels (etiquetas de los números en los ejes)
    ax.tick_params(axis='x', labelsize=16)
    ax.tick_params(axis='y', labelsize=16)

    plt.show()

```
